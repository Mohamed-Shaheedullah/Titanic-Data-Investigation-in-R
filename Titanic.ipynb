# Titanic data set in R

## Import data

```r
ttnc_data <- read.csv(file='//home//raf//Dropbox//DropBoxDocuments//Learning//Intequal//R_stats//Titanic//titanic.csv', header = T)
```

### Make necessary factors

```r
# make factors for survived, sex and class
ttnc_data$Survived = as.factor(ttnc_data$Survived)
ttnc_data$Pclass = as.factor(ttnc_data$Pclass)
ttnc_data$Sex = as.factor(ttnc_data$Sex)
```

```r
str(ttnc_data)
```

## Logistic regression

```r
# try model
model_logistic2 = glm(Survived~Pclass+Sex,na.action=na.omit, family = binomial , data = ttnc_data )
```

```r
summary(model_logistic2)
```

## Charts and graphs

### survival rate basic

```r
barplot(table(ttnc_data$Survived), main='Titanic survival rate')
print(table(ttnc_data$Survived))
```

### plot for gender

```r
surv_by_gender <- table(ttnc_data$Survived, ttnc_data$Sex)
```

```r
barplot(surv_by_gender, main='Titanic survival rate by gender', xlab='Gender', ylab='Number of Passengers', col = c('darkblue', 'red'), legend=rownames(surv_by_gender))
```

```r
# show numbers and proportions

surv_by_gender

print(prop.table(surv_by_gender, margin =2))
```

### plot for class

```r
surv_by_class <- table(ttnc_data$Survived, ttnc_data$Pclass)
```

```r
barplot(surv_by_class, main='Titanic survival rate by class', xlab='class', ylab='Number of Passengers', col = c('darkblue', 'red'), legend=rownames(surv_by_class))
```

## Age and Gender

### subset the data by age

```r
under30_data <- subset(ttnc_data, Age < 30)
between_30_50 <- subset(ttnc_data, Age >= 30 & Age <=50)
over50_data <- subset(ttnc_data, Age > 50)
```

```r
surv_by_under30_gender <- table(under30_data$Survived, under30_data$Sex)
```

```r
barplot(surv_by_under30_gender, main='Titanic survival rate by Under 30 gender', xlab='Gender', ylab='Number of Passengers', col = c('darkblue', 'red'), legend=rownames(surv_by_under30_gender))
```

```r
surv_by_between_30_50 <- table(between_30_50$Survived, between_30_50$Sex)
```

```r
barplot(surv_by_between_30_50, main='Titanic survival rate between 30 to 50 by gender', xlab='Gender', ylab='Number of Passengers', col = c('darkblue', 'red'), legend=rownames(surv_by_between_30_50))
```

```r
surv_by_over_50_gender <- table(over50_data$Survived, over50_data$Sex)
```

```r
barplot(surv_by_over_50_gender, main='Titanic survival rate over 50 by gender', xlab='Gender', ylab='Number of Passengers', col = c('darkblue', 'red'), legend=rownames(surv_by_over_50_gender))
```
